```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img= np.array(Image.open("images/tree.png").convert("L"))

plt.imshow(img,cmap= "gray")

```

```python
W,H,=img.shape
print(img.shape)
```

```python
message = input()
message += "[END]"
message = message.encode("ascii")
message_bits = ' '.join([format(i, '08b') for i in message])  # Corrected format usage

print(message_bits)

```

```python
img =img.flatten()

for idx,bit in enumerate(message_bits):
    val = img[idx]
    val = format(val, '08b')
    val = val[:-1]+bit
    img[idx] = int(val,2)
img = img.reshape((W,H))
```

```python
plt.imshow(img, cmap="gray")
```

```python
msg = " "
idx = 0
img = img.flatten()

max_iterations = img.size // 8  # Limit to avoid infinite loop

while msg[-5:] != '[END]' and idx < max_iterations:
    bits = ''.join(format(i, '08b')[-1] for i in img[idx:idx + 8])
    
    try:
        msg += chr(int(bits, 2))
    except ValueError:
        msg += '?'  # Handle decoding errors gracefully

    idx += 8  

    if idx > img.size - 8:
        print("There is no hidden message.")
        break

print("Hidden message:", msg)
```
